# <i>Tokenization</i>

### Toenization is subdivided into 3 different parts:
#### 1. Trigams: tokens of 3 consecutive written words.
#### 2. Bigrams: tokens of 2 consecutive written words.
#### 3. Ngrams: tokens of ANY nuimber of consecutive written words.

```python
string = "La costiera, formata dal deposito di tre grossi torrenti, il Fiumelatte, il Pioverna, e il Troggia, sbocca qui nella vasta pianura, e, declinando leggermente, fino alle rive stesse del lago, coperta di vigne e di olivi, con ville sparse di tratto in tratto, offre un aspetto gaio e ridente, anche al più malinconico viaggiatore."
```

```python
quotes_tokens = nltk.word_tokenize(string) # create the tokens
quotes_tokens
```

```python
quote_bigrams = list(nltk.bigrams(quotes_tokens))
quote_bigrams



# These create bigram tokens in the 2nd paragraph of Promessi Sposi
```

```python
quote_trigrams = list(nltk.trigrams(quotes_tokens))
quote_trigrams

# These create trigram tokens in the 2nd paragraph of Promessi Sposi
```

```python
quote_Ngrams = list(nltk.ngrams(quotes_tokens, 5))
quote_Ngrams

# This creates Ngram tokens in the 2nd paragraph of Promessi Sposi. the "5" is arbitrary.
```

# Stemming
### Normalizes the words into their base or root form.
### Example: Affection, affected, affectionate -> all come from the root word "Affect"

```python
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer # Importer stemmer modules

pst = PorterStemmer()
pst.stem("having") # PorterStemmer the word having to its root word, have.
```

```python
# Now let's figure out how to STEM an entire list of words.
wordlist = ["wording", "giving", "listing"]

for word in wordlist:
    print(f"Stemming for '{word}' is: '{pst.stem(word)}'.")
```

```python
lst = LancasterStemmer()

for word in wordlist:
    print(f"Stemming for '{word}' is: '{lst.stem(word)}'.") # This is more aggressive that Porter
```

```python
sst = SnowballStemmer('english') # With SnowballStemmer you need to define the language used.

for word in wordlist:
    print(f"Stemming for '{word}' is: '{sst.stem(word)}'.") # This is more aggressive that Porter
```

# Parts of Speech 

### Indicated how a qword functions in meaning, grammatically, wihtin a sentence. Divided between Tags and Description
### Tags: Used to describe whether a word is a noun or adjective, pronoun, symbol ect.

### Example: "The waiter cleared the plates from the table".
1. THE =  Determiner (DT)
2. WAITER =  Noun (NN)
3. CLEARED =  Past Tense Verb (VBD)
4. THE =  Determiner (DT)
5. PLATES =  Noun (NN)
6. FROM =  N/A (not defined)
7. THE =  Determiner (DT)
8. TABLE =  Noun (NN)





```python
# Output Tags of every token within a sentence

string1 = "Simon is a nautral when it comes to drawing"

string1_tokens = word_tokenize(string1)

for token in string1_tokens:
    print(nltk.pos_tag([token]))
```

# NER: Named Entity Recognition

### NER Entities List: 
1. Facility  2. Location  3. Organization  4. Person  5. Geo-Political Entity  6. Geo-Socio-Political Group

### The process of detecting a NER has 3 phases:
1. Noun Phase Identification: Dependency Parsing, and Partial Speech Tagging
2. Phase Classification: all phased Nouns are classified into NERs
3. Entity Disambiguation

#### Example: Google's CEO SUndar Pichai introduced the new pixel at Minnesota Roi Centre Event
#### Google = Organization; Sundar Pichai = Person; Minnesota = Location; Roi Centre Event = Organization

```python
from nltk import ne_chunk

string2 = "The US President lives in the White House"

string2_tokens = word_tokenize(string2)
string2_tags = nltk.pos_tag(string2_tokens)
string2_tags
```

```python
string2_pos = ne_chunk(string2_tags)
print(string2_pos)
```

# <i>Chunking</i>
### Picking up individual pieaces of information and Grouping them into bigger Pieces.

#### Example: "We caught the pink panther"
#### We = Verb, Caught = Verb, The Pink Panther = Noun Chunk


```python
string3 = "The big cat ate the little mouse who was after the fresh cheese"

string3_tokens = ne_chunk(nltk.pos_tag(word_tokenize(string3))) # chunking in a single code line

print(string3_tokens)
```

```python
# Output the Syntax tree of String 3, to do this you need to install the SVGLING python module

grammar_np = r"NP: {<DT>?<33>*<NN>}"

chunk_parser = nltk.RegexpParser(grammar_np)

chunk_result = chunk_parser.parse(string3_tokens)
chunk_result
```
